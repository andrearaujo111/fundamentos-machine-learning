# 1. IMPORTS

```python
import sklearn.datasets as datasets
import sklearn.cluster as cluster
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
```

# 2. CREATE DATA

```python
# Create datasets
X, y = datasets.make_blobs(n_samples=3200, n_features = 2, centers=3, cluster_std=1.5, random_state=0)

# Plot data
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

# 3.0 TRAIN DATA

```python
# Define model
k_means = cluster.KMeans(n_clusters=3, init='random', n_init=10, random_state=0)

# Fit
k_means.fit(X)

# Predict
clusters = k_means.predict(X)

# Plot the result visually
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.show()
```

```python
# Checking silhouette score
ss_average = metrics.silhouette_score(X, clusters)
print(f'The average silhouette score is: {ss_average}')
```

# 4.0 FIND THE BEST VALUES FOR K (ELBOW METHOD) - HIGH VARIATION

```python
# Create new dataset
X, y = datasets.make_blobs(n_samples=1200, n_features = 2, centers=6, cluster_std=2.23, random_state=0)

# Plot data
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
# Create a range that will iterate over the algorithm
n_clusters = range(2, 12, 1)

# Create an empty list to save the performance
ss_results = []

for i in n_clusters:

    # Define
    kmeans = cluster.KMeans(n_clusters=i, init='random', n_init=10, random_state=0)

    # Train
    kmeans.fit(X)

    # Predict
    k_performance = kmeans.predict(X)

    # Performance
    ss_score = metrics.silhouette_score(X, k_performance)
    ss_results.append(ss_score)

# Plot the result
plt.plot(n_clusters, ss_results, marker='o')
plt.xlabel('n_clusters')
plt.ylabel('silhouette_score')
plt.show()
```

```python
# Define
kmeans = cluster.KMeans(n_clusters=2, init='random', n_init=10, random_state=0)

# Train
kmeans.fit(X)

# Predict
labels = kmeans.predict(X)

# Performance
ss_score = metrics.silhouette_score(X, labels)

# Plot the result
print(f'The silhouette score is: {ss_score}')
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```
