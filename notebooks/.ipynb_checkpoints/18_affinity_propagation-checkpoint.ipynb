## 0. IMPORTS

```python
import warnings
import time
import numpy             as np
import IPython.display   as display
import sklearn.cluster   as cluster
import sklearn.datasets  as datasets
import sklearn.metrics   as metrics
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
```

## 1. USING AFFINITY PROPAGATION WITH CLASS' EXAMPLE

```python
# create dataset
X = np.array ([
    [3, 4, 3, 2, 1],
    [4, 3, 5, 1, 1],
    [3, 5, 3, 3, 3],
    [2, 1, 3, 3, 2],
    [1, 1, 3, 2, 3]
])

# define model
model = cluster.AffinityPropagation()

# train
model.fit(X)

# predict
labels = model.predict(X)

# show labels
labels
```

## 2. TRAINING AFFINITY PROPAGATION WITH SYNTHETIC DATA

```python
# create dataset
X, _ = datasets.make_blobs(n_samples=200, centers=7, n_features=2, random_state=0, cluster_std=0.6)

# show data behavior
plt.scatter(X[:, 0], X[:, 1], alpha=0.6, edgecolors='r')
plt.show()
```

```python
# model definition
af2 = cluster.AffinityPropagation(preference=-40)

# model training
af2.fit(X)

# predict
labels = af2.predict(X)

# show clusters
#fig = plt.figure()
plt.title('Clustering')
plt.scatter(X[:, 0], X[:, 1], alpha=0.6, c=labels, edgecolors='r')
plt.show()
```

```python
# how many clusters there is:
len(np.unique(labels))
```

## 3. ALGORITHM CONVERGENCE ON VISUAL MODEL

```python
# create dataset
X, _ = datasets.make_blobs(n_samples=600, centers=4, cluster_std=0.4, random_state=0)

# shpw figure
plt.scatter(X[:, 0], X[:, 1], alpha=0.6, edgecolor='b')
plt.show()
```

```python
# create a range that will serve as parameters
p = range(-60, 61)

# empty list to save silhouette score results
s_score = []

fig = plt.figure()

# test parameters
for i in range (len(p)):
    ap3 = cluster.AffinityPropagation(preference=p[i])
    ap3.fit(X)
    labels = ap3.predict(X)

    # plot
    plt.scatter(X[:, 0], X[:, 1], alpha=0.6, c=labels, cmap='rainbow', edgecolor='b')
    display.clear_output(wait=True)
    plt.show()
    plt.gcf()
    time.sleep(0.5)

    print(f'Iterao: {i}')
    print(f'Valor de p: {p[i]}')
    print(f'N clusters {len(np.unique(labels))}')
```

## 4. ALGORITHM CONVERGENCE WITH PERFORMANCE METRIC

```python
# create dataset
X, _ = datasets.make_blobs(n_samples=300, centers=6, n_features=2, random_state=0, cluster_std=0.7)

# show data behavior
plt.scatter(X[:, 0], X[:, 1], alpha=0.6, edgecolors='r')
plt.show()
```

```python
# create a range that will serve as parameters
p = np.arange(-1, -61, -1)

# empty list to save silhouette score results
s_list = []

fig = plt.figure()

# test parameters
for i in range (len(p)):
    ap4 = cluster.AffinityPropagation(preference=p[i])
    ap4.fit(X)
    labels = ap4.predict(X)

    # performance
    s_score = metrics.silhouette_score(X, labels)
    s_list.append(s_score)

    # plot
    plt.plot(p[:i], s_list[:i], marker='o', c='red');
    plt.title('Affinity Propagation - Convergence')
    plt.xlabel('Preference')
    plt.ylabel('Silhouette Score')
    
    display.clear_output(wait=True)
    plt.show()
    plt.gcf()
    time.sleep(0.5)

    print(f'Iterao: {i}')
    print(f'Valor de p: {p[i]}')
    print(f'N clusters {len(np.unique(labels))}')
```

```python
# final clustering
max_index = s_list.index(max(s_list))
af = cluster.AffinityPropagation(preference=p[max_index])
clustering = af.fit(X)
fig = plt.figure()
plt.title( "Separao dos Clients VIP")
plt.scatter(X[:,0], X[:,1], c=clustering.labels_, cmap='rainbow', alpha=0.7, edgecolors='b');
```
