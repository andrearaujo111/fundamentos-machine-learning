# 1. IMPORTS

```python
import sklearn.datasets as datasets
import sklearn.cluster as cluster
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
```

# 2. TRAIN K-MEANS - 2 FEATURES - LOW VARIABILITY

```python
# Create syntethic data
X, y = datasets.make_blobs(n_samples=500, n_features=2, centers=3, cluster_std=0.38, random_state=0)

# Define
model =  cluster.KMeans(n_clusters=3, init='random', n_init=15, random_state=0)

# Fit + Predict
labels = model.fit_predict(X)

# Performance
ss_avg = metrics.silhouette_score(X, labels)
ss_avg 
```

```python
# Find the best value for K
values = list(range(2, 12))
ss_scores = []

# Scroll through the different K values, train the model and save the performance for each k value
for i in values:
    # Define
    model =  cluster.KMeans(n_clusters=i, init='random', n_init=10, random_state=0)

    # Fit 
    model.fit(X)
    
    # Predict
    labels = model.predict(X)

    # Performance
    ss_avg = metrics.silhouette_score(X, labels)
    ss_scores.append(ss_avg)

# Plot the result
plt.plot(values, ss_scores, marker='o', linewidth=2)
plt.show()
```

# 3. TRAIN K-MEANS - 8 FEATURES - HIGH VARIABILITY

```python
# Create syntethic data
X, y = datasets.make_blobs(n_samples=500, n_features=8, centers=3, cluster_std=2.43, random_state=0)

# Define
model =  cluster.KMeans(n_clusters=3, init='random', n_init=10, random_state=0)

# Fit + Predict
labels = model.fit_predict(X)

# Performance
ss_avg = metrics.silhouette_score(X, labels)
ss_avg 
```

The idea here is to show how the performance is bad when there's high variability, and that still k-means will bring a cluster definition for us.

# 5. CREATE NON-SPHERICAL CLUSTERS

```python
# Create data
X, y = datasets.make_moons(n_samples=500, noise=0.05, random_state=0)

# Plot data
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# Find the best value for K
values = list(range(2, 20))
ss_scores = []

# Scroll through the different K values, train the model and save the performance for each k value
for i in values:
    # Define
    model =  cluster.KMeans(n_clusters=i, init='random', n_init=10, random_state=0)

    # Fit 
    model.fit(X)
    
    # Predict
    labels = model.predict(X)

    # Performance
    ss_avg = metrics.silhouette_score(X, labels)
    ss_scores.append(ss_avg)

# Plot the result
plt.plot(values, ss_scores, marker='o', linewidth=2)
plt.show()
```

We can see that, instead of reaching a peak and starting to decrease, for non-spherical clusters find the best k as the k value increases

```python
# Train the model with the best k value

# Define
model =  cluster.KMeans(n_clusters=7, init='random', n_init=10, random_state=0)

# Fit 
model.fit(X)
    
# Predict
labels = model.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```
