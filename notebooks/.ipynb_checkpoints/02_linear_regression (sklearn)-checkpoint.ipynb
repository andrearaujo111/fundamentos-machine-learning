# 0. IMPORTS

```python
import pandas          as pd
import numpy           as np
import seaborn         as sns
import sklearn.metrics as metrics

from sklearn.linear_model import LinearRegression
from matplotlib           import pyplot as plt
```

# 1. Load Data

```python
# Load dataset
df = pd.read_csv('../datasets/train.csv')
df.head()
```

# 2. Prepare Data

```python
# Evaluate correlation
plt.figure(figsize=(14, 8))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
# Selecting features
features = ['idade', 'divida_atual', 'renda_anual', 'valor_em_investimentos',
'taxa_utilizacao_credito', 'num_emprestimos', 'num_contas_bancarias', 'num_cartoes_credito',
'dias_atraso_dt_venc', 'num_pgtos_atrasados', 'num_consultas_credito', 'taxa_juros']

# features2 = ['divida_atual', 'taxa_utilizacao_credito', 'dias_atraso_dt_venc']

# Selecting response variable
target = 'saldo_atual'

# Creating train datasets
x_train = df.loc[:, features]
y_train = df.loc[:, target]
```

# 3. Train Model

```python
# Create Linear Model object
lr_model = LinearRegression()

# Train the model
lr_model.fit(x_train, y_train)

# Make the predictions
y_pred = lr_model.predict(x_train)
```

# 4. Results

```python
# Creating new dataframe
df1 = df[['id_cliente', 'saldo_atual']].copy()

# Adding predictions to dataset, in order to compare results
df1['predicted_saldo'] = y_pred

# Calculating error
df1['error'] = df1['predicted_saldo'] - df1['saldo_atual']

df1.head()
```

```python
plt.figure(figsize=(10, 5))

ax = sns.histplot(data=df1, x='error', bins=20, kde=True)
plt.title('Error Distribution')
plt.show()
```

# 5. Performance

```python
r2 = metrics.r2_score(y_train, y_pred)
print(f'The R² score is: {r2}')
```

Com todas as features: **0.1692**  
Com somente as features ['divida_atual', 'taxa_utilizacao_credito', 'dias_atraso_dt_venc']: **0.1686**  

Vamos escolher todas as features pois a performance foi ligeiramente maior.
