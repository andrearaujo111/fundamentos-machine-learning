# 1. IMPORTS

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# 2. TRAINING ALGORITHM WITH 2 DEGREES

```python
# initialize random number generator 
np.random.seed(0)

m = 150

# create data
X = np.linspace(-10, 10, num=m)
y = 0.1*X**2 + X + 3 + np.random.normal(size=m)
```

```python
# plot distribution
plt.scatter(X, y)
plt.show()
```

```python
# Pre process data to include polynomials
poly = PolynomialFeatures(degree=2, include_bias=True)
X_poly = poly.fit_transform(X.reshape(-1, 1))

# Define 
lr = LinearRegression()

# Train
lr.fit(X_poly, y)

# predict
y_hat = lr.predict(X_poly)

# performance
r2 = r2_score(y, y_hat)
rmse = np.sqrt(mean_squared_error(y, y_hat))

print(f'R2: {r2}')
print(f'RMSE: {rmse}')
```

```python
# plot
plt.scatter(X, y)
plt.plot(X, y_hat, color='red')
plt.show()
```

# 3. TRAIN ALGORITHMS WITH MORE THAN 2 DEGREES

```python
# initialize random number generator 
np.random.seed(0)

m = 150

# create data
X = np.linspace(-11, 1, num=m)
y = 0.1*X**2 + X + 3 + np.random.normal(size=m)
```

```python
# plot distribution
plt.scatter(X, y)
plt.show()
```

```python
# define a set of polynomials
degrees = [i for i in range(1, 15)]
preds = []
r2_scores = []
rmse_scores = []

for i in degrees:
    
    # Pre process data to include polynomials
    poly = PolynomialFeatures(degree=i, include_bias=True)
    X_poly = poly.fit_transform(X.reshape(-1, 1))
    
    # Define 
    lr = LinearRegression()
    
    # Train
    lr.fit(X_poly, y)
    
    # predict
    y_hat = lr.predict(X_poly)

    # performance
    r2 = r2_score(y, y_hat)
    rmse = np.sqrt(mean_squared_error(y, y_hat))
    
    # append
    preds.append(y_hat)
    r2_scores.append(r2)
    rmse_scores.append(rmse)
```

```python
# plots
plt.scatter(X, y)
plt.plot( X, preds[0], color='red' ) # dg=1
plt.plot( X, preds[1], color='green' ) # dg=2
plt.plot( X, preds[2], color='black' ) # dg=3
plt.show()
```

```python
# plots
plt.scatter(X, y)
plt.plot( X, preds[4], color='yellow' ) # dg=5
plt.plot( X, preds[8], color='orange' ) # dg=9
plt.plot( X, preds[13], color='purple' ) # dg=14
plt.show()
```
