# 1. IMPORTS

```python
import sklearn.datasets as datasets
import sklearn.metrics as metrics
import sklearn.model_selection as ms
import sklearn.linear_model as lm

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from collections import Counter
from sklearn.neighbors import KNeighborsClassifier
```

# 2. CREATE SYNTETHIC DATA

```python
X, y = datasets.make_classification(n_samples=1000, n_classes=2, random_state=1)
```

# 3. PRECISION/RECALL CURVE WITH SYNTETHIC DATA

```python
# train/test/validation split
X_train, X_test, y_train, y_test = ms.train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
# Define 
model = lm.LogisticRegression(solver='lbfgs')

# Fit
model.fit(X_train, y_train)

# Predict probabilities
yhat_test = model.predict_proba(X_test)
yhat_test = yhat_test[:, 1]
yhat = model.predict(X_test)

# Predict class values
precision, recall, thresholds = metrics.precision_recall_curve(y_test, yhat_test)
f1 = metrics.f1_score(y_test, yhat)
print(f'Model f1 score: {f1}')
```

```python
# Showing the first value for precision (y), recall (x) and the first threshold
print(precision[0])
print(recall[0])
print(thresholds[0])
```

```python
# plot precision and recall curve
plt.plot(precision, recall, label='model')
plt.ylabel('precision')
plt.xlabel('recall')
plt.legend()
plt.show()
```

```python
# Choosing the best value
plt.plot(thresholds, precision[:-1], c='r')
plt.plot(thresholds, recall[:-1], c='b')
plt.vlines(x=0.535, ymin=0, ymax=1, linestyles='dashed')
plt.xlabel('thresholds')
```

```python
# check the number of values predicted in yhat (no probability)
Counter(yhat)
```

```python
# check the number of values predicted in yhat_test (probability)
pred_classes = [1 if i >= 0.535 else 0 for i in yhat_test]
Counter(pred_classes)
```

```python
# check new f1 score
metrics.f1_score(y_test, pred_classes)
```
